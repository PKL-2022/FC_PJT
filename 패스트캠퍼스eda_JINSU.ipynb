```python
# 구매완료한 강의만 
df = df[df.state=='COMPLETED']
# 데이터 사이언스 == 데이터사이언스 로 바꾸기
df.loc[df.category_title=='데이터 사이언스', 'category_title']= '데이터사이언스'
# payment만 남기고 나머지 전처리
df = df[df.type=='PAYMENT']
```

```python
df.category_title.unique()
```

```python
# 0원인 강의만 전처리
df = df[df.transaction_amount!=0]
# 10000원 이하인 강의만 전처리
df = df[df.transaction_amount>10000]
```

```python
df.describe()
```

```python
df.type.unique()
```

```python
# 카테고리 별 강의 개수 
# df.category_title.value_counts()

category_title = df.category_title.value_counts().keys()
values = df.category_title.value_counts().values

ratio = values
labels = category_title
explode = [0.05]*len(df.category_title.value_counts().keys())

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

```python
# 카테고리 별 금액
# df.groupby('category_title').sum().loc[:,'transaction_amount'].sort_values(ascending=False)

x = np.arange(len(df.category_title.value_counts().keys()))
category_title = df.groupby('category_title').sum().loc[:,'transaction_amount'].sort_values(ascending=False).keys()
values = df.groupby('category_title').sum().loc[:,'transaction_amount'].sort_values(ascending=False).values

plt.bar(x, values)
plt.xticks(x, category_title)

plt.show()
```

```python
# 카테고리 별 판매 대비 금액
category_course_money = (df.groupby('category_title').sum().loc[:,'transaction_amount']//df.category_title.value_counts()).sort_values(ascending=False)
category_course_money

x = np.arange(len(df.category_title.value_counts().keys()))
category_title = category_course_money.keys()
values = category_course_money.values

plt.bar(x, values)
plt.xticks(x, category_title)

plt.show()
```

# 카테고리에 따라서 쿠폰없이도 결재율이 어떻게 되는지

```python
# 쿠폰 적용 없는 강의 구매자
null_coupon = df[df.coupon_title.isnull()]
coupon = df[df.coupon_title.notnull()]
```

```python
print('쿠폰 없이 결제:',len(null_coupon))
print('쿠폰 결제:',len(coupon))
print('전체 결제:',len(df))
```

```python
null_coupon.describe()
```

```python
null_coupon.head(1)
```

```python
null_coupon.category_title.unique()
```

```python
# 쿠폰 없이 결제 수
null_coupon.groupby('category_title').count().loc[:,'customer_id'].sort_values(ascending=False)
```

```python
# 쿠폰 결제 수
coupon.groupby('category_title').count().loc[:,'customer_id'].sort_values(ascending=False)
```

```python
category = []
for i in null_coupon.category_title.unique():
    if i not in coupon.category_title.unique():
        category.append(i)
print('쿠폰 없이만 결제 가능한 카테고리: ',category)
```

```python
# 쿠폰 없이만 결제 가능한 카테고리 제외
null_coupon = null_coupon[((null_coupon['category_title'] != '교육')
             &(null_coupon['category_title'] != '관리')) 
            &(null_coupon['category_title'] != '지식콘텐츠')
            & (null_coupon['category_title'] != 'nan')
            & (null_coupon['category_title'] != '파이낸스')
           ]
```

```python
labels = coupon.category_title.unique()
null = (null_coupon.groupby('category_title').count().loc[:,'customer_id'].sort_values(ascending=False)).values
notnull = (coupon.groupby('category_title').count().loc[:,'customer_id'].sort_values(ascending=False)).values

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, null, width, label='null coupon')
rects2 = ax.bar(x + width/2, notnull, width, label='use coupon')

ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```
