### **Question**

- 할인율이 구매 결정에 중요한 요인일까?
- 기존 할인율이 높았을 경우, 재구매로 이어질까?
- 상위 20%의 할인을 받았던 그룹의 재구매율과 하위 20%의 할인을 받았던 그룹의 재구매율은 어떨까?
- 재구매 시, 80%에 영향을 미치는 20%는 상위 그룹일까 하위 그룹일까 또는 다른 그룹일까?

파레토법칙(Pareto Law)은 전체 결과의 80%가 전체 원인의 20%에서 일어나는 현상을 의미한다. 일례로, 인사행정에 있어서 공무원 20%가 전체성과의 80% 정도를 차지하는 것 등이 이에 해당된다. 대개 80대 20의 법칙, 8대 2의 법칙라고도 한다.

- '웰컴'/'신규'의 기간제 쿠폰을 사용한 결제한 경우, 재구매에 영향을 미칠까?

```python
df1 = df.copy()
```

```python
# case1 : 확정 결제 고객
case1 = (df1['type'] == 'PAYMENT') & (df1['state'] == 'COMPLETED')
case1.value_counts()
```

```python
# case2 : 취소 고객
case2 = (df1['type'] == 'PAYMENT') & (df1['state'] == 'CANCELLED')
case2.value_counts()
```

```python
# case3 : 확정 결제 고객 & 쿠폰 할인 적용 고객
case3 = (df1['type'] == 'PAYMENT') & (df1['state'] == 'COMPLETED') & (df1['coupon_discount_amount'] > 0)
case3.value_counts()
```

```python
# 쿠폰명 결측치 없음 확인 / 강사 지인 찬스 포함
df1.loc[case1 & case3, 'coupon_title'].value_counts().sum()
```

```python
### case4 : 확정 결제 고객 & 쿠폰 할인 적용 고객 & 실제 결제 금액이 0을 초과하는 고객 > 이상치 제거1
case4 = case3 & (df1['transaction_amount'] > 0)
sorted_case = df1.loc[case3, ['transaction_amount']] > 0
sorted_case.value_counts()
```

```python
# coupon_aver : 쿠폰 할인 평균 비용
coupon_aver = (round(df1.loc[case4, 'coupon_discount_amount'].mean(),2))
coupon_aver
```

```python
# coupon_perc : 쿠폰 할인 %
coupon_perc = df1['discount'] = round(df1['coupon_discount_amount'] / df1['sale_price'] * 100, 2)
coupon_perc.value_counts()
```

```python
# discount column 수 재 확인
df1.loc[case4, 'discount'].count()
```

```python
# column명 재 확인
df1.columns
```

```python
# discount 수 더블 체크 / 필요 column 정보 추출
df2 = df1.loc[case4, ['customer_id', 'type', 'state','course_title', 'format', 'coupon_title', 'transaction_amount', 'coupon_discount_amount', 'sale_price','discount']]
df2
```

```python
# 할인비율 100% 초과, 이상치 3항목 확인
df2 = df2.sort_values(by = ['discount'], ascending = False)
df2
```

```python
# 할인비율 100% 초과, 이상치 3항목 삭제 > 이상치 제거2
final_case = df2.drop(labels = [99824, 91203, 89739], axis = 0)
final_case
```

```python
# 전처리 완료 / 중복인원 제거는 추후 예정
df2 = final_case.reset_index()
df2
```

```python
# 할인율 분포 확인
bins = [(df2['discount'].min()), 20, 40, 60, 80, (df2['discount'].max())]
test_cut = pd.cut(df2['discount'], bins, right=False).value_counts()
test_cut
```

```python
# 할인율 분포 시각화
plt.title('discount rate distribution', fontsize = 15)

plt.xlabel('discount rate', fontsize = 10)
plt.ylabel('frequency', fontsize = 10)

plt.hist(df2['discount'], bins = bins, color='B', alpha=0.7)
plt.show()
```
