```python
# 구매완료한 강의만 
df = df[df.state=='COMPLETED']
# 데이터 사이언스 == 데이터사이언스 로 바꾸기
df.loc[df.category_title=='데이터 사이언스', 'category_title']= '데이터사이언스'
# payment만 남기고 나머지 전처리
df = df[df.type=='PAYMENT']
# 0원인 강의만 전처리
df = df[df.transaction_amount!=0]
# 10000원 이하인 강의만 전처리
# df = df[df.transaction_amount>10000]
# category_title nan 값 제거 3개
df = df.loc[~(df['category_title'].isnull())]
```

```python
# course_title, format 'B2B' 전처리
df = df.loc[~(df['course_title'].str.contains('B2B')|(df['format'].str.contains('B2B')))]
# 'B2B' 전처리 확인
df.loc[(df['course_title'].str.contains('B2B')|(df['format'].str.contains('B2B')))]
```

```python
# 개발자 테스트용 3건 전처리
df = df[~(df['course_title'].str.contains('TEST|test'))]
# 기간 한정판매 강의 100건 전처리
df = df[~df['course_title'].str.contains('한정판매')]
```

```python
# 상위 10개 카테고리 별 강의 비율
# df.category_title.value_counts()

plt.figure(figsize=(8,8))

category_title = df.category_title.value_counts().head(10).keys()
values = df.category_title.value_counts().head(10).values

ratio = values
labels = category_title
explode = [0.05]*len(df.category_title.value_counts().head(10).keys())

plt.title('상위 10개 카테고리 강의 비율', fontsize=15)

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

```python
# 카테고리 별 금액
# df.groupby('category_title').sum().loc[:,'transaction_amount'].sort_values(ascending=False)
plt.figure(figsize=(15,3))
x = np.arange(len(df.category_title.value_counts().keys()))
category_title = df.groupby('category_title').sum().loc[:,'transaction_amount'].sort_values(ascending=False).keys()
values = df.groupby('category_title').sum().loc[:,'transaction_amount'].sort_values(ascending=False).values

plt.bar(x, values)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.xticks(x, category_title)
plt.title('카테고리 별 총 매출', fontsize=15)

plt.show()
```

```python
# 카테고리 별 판매 대비 금액
category_course_money = (df.groupby('category_title').sum().loc[:,'transaction_amount']//df.category_title.value_counts()).sort_values(ascending=False)
category_course_money
plt.figure(figsize=(15,3))
x = np.arange(len(df.category_title.value_counts().keys()))
category_title = category_course_money.keys()
values = category_course_money.values

plt.bar(x, values)
plt.xticks(x, category_title)
plt.title('카테고리 별 강의 당 가격', fontsize=15)

plt.show()
```

# 카테고리에 따라서 쿠폰없이도 결재율이 어떻게 되는지

```python
# 쿠폰 적용 없는 강의 구매자
null_coupon = df[df.coupon_title.isnull()]
coupon = df[df.coupon_title.notnull()]
```

```python
print('쿠폰 없이 결제:',len(null_coupon))
print('쿠폰 결제:',len(coupon))
print('전체 결제:',len(df))
```

```python
# 쿠폰 없이 결제 수
null_coupon.groupby('category_title').count().loc[:,['id']].sort_values('id',ascending=False)
```

```python
# 쿠폰 결제 수
coupon.groupby('category_title').count().loc[:,['id']].sort_values('id', ascending=False)
```

```python
category = []
for i in null_coupon.category_title.unique():
    if i not in coupon.category_title.unique():
        category.append(i)
print('쿠폰 없이만 결제 가능한 카테고리: ',category)
```

```python
# 쿠폰 없이만 결제 가능한 카테고리 제외
null_coupon = null_coupon[~null_coupon['category_title'].str.contains('지식콘텐츠|파이낸스')]
```

```python
labels = coupon.category_title.unique()
null = (null_coupon.groupby('category_title').count().loc[:,'customer_id'].sort_values(ascending=False)).values
notnull = (coupon.groupby('category_title').count().loc[:,'customer_id'].sort_values(ascending=False)).values

x = np.arange(len(labels),)
width = 0.3

# plt.figure(figsize=(20,3))
fig, ax = plt.subplots()
rects1 = ax.bar(x- width/2, null, width, label='쿠폰 미사용')
rects2 = ax.bar(x+ width/2, notnull, width, label='쿠폰 사용')

ax.set_xticks(x, labels)


ax.bar_label(rects1, padding=5)
ax.bar_label(rects2, padding=5)

fig.tight_layout()
plt.show()
```

```python
# 쿠폰 상대도수 분포
relative_frequency = null_coupon.groupby('category_title').count().loc[:,'customer_id']/(null_coupon.groupby('category_title').count().loc[:,'customer_id']+coupon.groupby('category_title').count().loc[:,'customer_id'])
y = list(relative_frequency.keys())
x = relative_frequency.values

# 쿠폰 없이 구매 비율
pd.DataFrame(x,y,columns=['relative_frequency']).sort_values('relative_frequency',ascending=False)
```
