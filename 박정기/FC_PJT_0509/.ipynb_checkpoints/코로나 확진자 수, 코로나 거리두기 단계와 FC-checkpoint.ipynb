```python
import pandas as pd
```

```python
covid_df = pd.read_csv('covid_seoul.csv')
```

```python
covid_df = covid_df.drop('Unnamed: 0', axis = 1)
```

```python
covid_df
```

```python
FC_df = pd.read_excel('파이널프로젝트_RAW_210329_210926.xlsx')
```

```python
FC_df
```

```python
cond = (FC_df.type == 'PAYMENT') & (FC_df.state == 'COMPLETED')
```

```python
FC_df = FC_df[cond]
```

```python
FC_df = FC_df[['type', 'state', 'completed_at']]
```

```python
FC_df = FC_df.reset_index(drop = True)
```

```python
FC_df
```

```python
def rmv_time(df):
    temp = []
    for d in df['completed_at']:
        d = d[:10]
        temp.append(d)
    return temp
```

```python
temp_list = rmv_time(FC_df)
```

```python
FC_df['completed_at'] = pd.to_datetime(FC_df['completed_at'], unit='s')
```

```python
FC_df['completed'] = temp_list
```

```python
FC_df.drop('completed_at', axis = 1)
```

```python
cnt_df = FC_df.groupby('completed').size()
```

```python
cnt_df = pd.DataFrame(cnt_df, columns=['cnt'])
```

```python
cnt_df
```

```python
import seaborn as sns
```

```python
sns.lineplot(x='completed', y='cnt', data = cnt_df)
```

```python
covid_df
```

```python
sns.lineplot(x='서울시 기준일', y='서울시 추가 확진', data = covid_df)
```
