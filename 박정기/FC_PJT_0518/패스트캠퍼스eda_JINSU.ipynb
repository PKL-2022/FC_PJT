# 카테고리에 따라서 쿠폰없이도 결재율이 어떻게 되는지

```python
# 쿠폰 적용 없는 강의 구매자
null_coupon = df[df.coupon_title.isnull()]
coupon = df[df.coupon_title.notnull()]
```

```python
print('쿠폰 없이 결제:',len(null_coupon))
print('쿠폰 결제:',len(coupon))
print('전체 결제:',len(df))
```

```python
# 쿠폰 없이 결제 수
null_coupon.groupby('category_title').count().loc[:,['id']].sort_values('id',ascending=False)
```

```python
# 쿠폰 결제 수
coupon.groupby('category_title').count().loc[:,['id']].sort_values('id', ascending=False)
```

```python
category = []
for i in null_coupon.category_title.unique():
    if i not in coupon.category_title.unique():
        category.append(i)
print('쿠폰 없이만 결제 가능한 카테고리: ',category)
```

```python
# 쿠폰 없이만 결제 가능한 카테고리 제외
null_coupon = null_coupon[~null_coupon['category_title'].str.contains('지식콘텐츠|파이낸스')]
```

```python
labels = coupon.category_title.unique()
null = (null_coupon.groupby('category_title').count().loc[:,'customer_id'].sort_values(ascending=False)).values
notnull = (coupon.groupby('category_title').count().loc[:,'customer_id'].sort_values(ascending=False)).values

x = np.arange(len(labels),)
width = 0.3

# plt.figure(figsize=(20,3))
fig, ax = plt.subplots()
rects1 = ax.bar(x- width/2, null, width, label='쿠폰 미사용')
rects2 = ax.bar(x+ width/2, notnull, width, label='쿠폰 사용')

ax.set_xticks(x, labels)


ax.bar_label(rects1, padding=5)
ax.bar_label(rects2, padding=5)

fig.tight_layout()
plt.show()
```

```python
# 쿠폰 상대도수 분포
relative_frequency = null_coupon.groupby('category_title').count().loc[:,'customer_id']/(null_coupon.groupby('category_title').count().loc[:,'customer_id']+coupon.groupby('category_title').count().loc[:,'customer_id'])
y = list(relative_frequency.keys())
x = relative_frequency.values

# 쿠폰 없이 구매 비율
pd.DataFrame(x,y,columns=['relative_frequency']).sort_values('relative_frequency',ascending=False)
```
