```python
cond = (FC_df['format'] == '올인원')
```

```python
all_df = FC_df[cond]
```

```python
all_df
```

```python
# 올인원 총 매출
all_df['transaction_amount'].sum()
```

```python
# 올인원 카테고리별 결제 횟수
all_df.groupby('category_title').size().sort_values(ascending = False)
```

```python
a_list = all_df.groupby('category_title').size().index.tolist()
```

```python
bill_list = []
for l in a_list:
    print(l)
    cond = (FC_df['category_title'] == l)
    temp_df = FC_df[cond]
    print((temp_df['transaction_amount'].sum()))
    bill_list.append(temp_df['transaction_amount'].sum())
    
```

```python
bill_df = pd.DataFrame({'category':a_list,'bill':bill_list})
```

```python
bill_df
```

```python
# 올인원, 각각의 카테고리별 매출
bill_df.sort_values(by = 'bill', ascending = False)
```

```python
B = bill_df.sort_values(by = 'bill', ascending = False)
```

```python
B
```

```python
sns.barplot(data = B, x = B.category, y = B.bill)
```

```python
c = all_df.groupby('category_title').size().sort_values(ascending = False)
```

```python
M = pd.merge(left = B , right = C, how = "inner", on = "category")
```
